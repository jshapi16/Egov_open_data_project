```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
from pprint import pprint
import requests
import json
import html
from scipy.stats import linregress
import statsmodels.api as sm
import seaborn as sns
```

```python
csv_merge = "../Resources/newmerge_1416.csv"
df_1416 = pd.read_csv(csv_merge)
```

```python
df_1416.columns
```

```python
df_1416_readiness = df_1416[["Country","Region_2014","Readiness-Scaled_2014", "Readiness-Scaled_2016"]]
```

```python
df_1416_per = df_1416_readiness[["Readiness-Scaled_2014", "Readiness-Scaled_2016"]]
df_1416_per.head()
```

```python
percentage = df_1416_per.pct_change(axis=1)
percentage = round(percentage * 100,2)
percentage_2016 = percentage["Readiness-Scaled_2016"].values.tolist()
percentage_2016
```

```python
df_1416_readiness.insert(4, "Percentage Change", percentage_2016, True)
```

```python
df_1416_readiness
```

```python
df_1416_readiness["Difference"] = df_1416_readiness["Readiness-Scaled_2016"].sub(df_1416_readiness["Readiness-Scaled_2014"],axis=0)
```

```python
df_1416_readiness
```

```python
grouped = df_1416_readiness.groupby(["Region_2014"])
```

```python
print("Readiness Score: Mean Percentage Change")
mean = grouped["Percentage Change"].mean()
mean.reset_index()
```
