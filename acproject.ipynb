```python
# linear regression function
#calling this function, passing to it:
#1. the x-data
#2. the y-data
#3. the label for the x-axis
#4. the label for the y-axis
#5. the title of the plot

def linear_plot(x,y,xstring,ystring,thistitle):
    #calling lingress() method that was imported above
    res = linregress(x, y)
    slope = res.slope
    intercept = res.intercept
    
    #create the plot
    #first, plotting actual data
    plt.plot(x,y,'b.',label='Data')
    #plot linear regression fit
    plt.plot(x,intercept + slope*x, 'r-', label='Linear Fit')
    #finally, adding axis labels, title, and legend
    plt.xlabel(xstring)
    plt.ylabel(ystring)
    plt.title(thistitle)
    plt.legend()
    plt.show()
    
    
    #Does the y-variable actually depend on the x-variable or not?
    #
    #1. The slope of the linear regression, WITH AN ASSOCIATED UNCERTAINTY
    #2. the R^2 value (i.e. the correlation coefficient)
    #
    #Using the Student's t-distribution
    #
    from scipy.stats import t
    tinv = lambda p, dof: abs(t.ppf(p/2, dof))
    ts = tinv(0.05, len(x)-2)
    print(f"R-squared: {res.rvalue**2:.6f}")
    print(f"slope (95%): {res.slope:.6f} +/- {ts*res.stderr:.6f}")
```
