# ODB vs Human capital 2014

```python
x_axis = df_14_16['ODB-Score-Scaled_2014'] *100
y_axis = df_14_16['Human Capital Index_2014'] * 100

def weather_graphing(x_axis, y_axis, title, xlabel, ylabel, text_coordinates):
    #linear regression
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
    regress_values = x_axis * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.figure(figsize=(30,15))
    sns.scatterplot(x = x_axis, y = y_axis, data = df_14_16, hue = 'Region_2014')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.annotate(line_eq, text_coordinates,fontsize=15,color="red")
    plt.plot(x_axis,regress_values,"r-")
    
weather_graphing(x_axis, y_axis, "ODB vs Human capital, 2014", "ODB 2014", "Human Capital 2014", (3,14))
for i, label in enumerate(df_14_16['Country']):
    plt.text(x_axis[i], y_axis[i],label)
```
